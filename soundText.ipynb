# 音声をテキストに変換します

## 以下のセルには、音声データをテキストに変換するPythonコードが含まれています。
音声データは、スマホなどで録音したものや配布されたデータを使用しても構いません。
小さなデータを使用すると、処理が早く終わります。

```python
# 音声データをテキストにします。音声データをアップロードします。
# ステップ1：必要なライブラリのインストール
!pip install git+https://github.com/openai/whisper.git
!pip install soundfile

# ステップ2：ライブラリのインポート
import whisper
import os
from google.colab import files

# ステップ3：Whisperモデルのロード
# model = whisper.load_model("base")
model = whisper.load_model("large")

# ステップ4：音声ファイルのアップロード
uploaded = files.upload()
audio_file = next(iter(uploaded))

# ステップ5：音声データのテキスト化
result = model.transcribe(audio_file, language='ja')
print("認識結果：")
print(result["text"])
```

```python
## 複数ファイル
# ステップ1：必要なライブラリのインストール
!pip install git+https://github.com/openai/whisper.git
!pip install soundfile

# ステップ2：ライブラリのインポート
import whisper
import os
from google.colab import files

# ステップ3：Whisperモデルのロード
# model = whisper.load_model("base")
model = whisper.load_model("large")

# ステップ4：音声ファイルのアップロード
print("音声ファイルを選択してください。複数選択も可能です。")
uploaded = files.upload()

# ステップ5：音声データのテキスト化（繰り返し処理）
for audio_file in uploaded.keys():
    print(f"\nファイル名: {audio_file}")
    result = model.transcribe(audio_file, language='ja')
    print("認識結果：")
    print(result["text"])
```

# テキストを音声に変換します

## ここではテキストを音声に変換する方法を学びます。

Google Colabではインストールできる音声処理ライブラリが限られているため、
音声合成には、edge-ttsライブラリを使用します。

```python
## テキストを音声に
```

```python
# ステップ1：必要なライブラリのインストール
!pip install gTTS

# ステップ2：ライブラリのインポート
from gtts import gTTS
from IPython.display import Audio

# ステップ3：テキストの入力
text = "こんにちは、これはテキストから音声への変換デモです。"

# ステップ4：音声データの生成
tts = gTTS(text=text, lang='ja')
tts.save("output.mp3")

# ステップ5：音声の再生
Audio("output.mp3", autoplay=True)
```

```python
import edge_tts
import asyncio
from IPython.display import Audio

async def tts():
    text = "こんにちは、これは音声の速度と声を変更するデモです。"
    # voice = "ja-JP-NanamiNeural"  # 日本語の女性の声
    voice = "ja-JP-KeitaNeural"  # 男性の声
    rate = "+10%"  # 速度を10%速くする
    output_file = "output.mp3"

    communicate = edge_tts.Communicate(text=text, voice=voice, rate=rate)
    await communicate.save(output_file)

# 非同期関数を実行
await tts()

# 音声の再生
Audio("output.mp3", autoplay=True)
```
